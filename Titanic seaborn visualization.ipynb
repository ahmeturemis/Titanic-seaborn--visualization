```python
import pandas as pd
titanic = pd.read_csv("train.csv")
col = ["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
titanic = titanic[col].dropna() #choose columns from col list and drop missing value rows
print(titanic.head(3))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline   
sns.distplot(titanic["Age"]) # visualize the distribution of the "Age" column
plt.show()
```

```python
sns.kdeplot(titanic["Age"],shade=True) #generate just the kernel density plot and shade under line
plt.xlabel("Age") #label bottom
sns.set_style("darkgrid") #other styles are "whitegrid" "dark" "white" "ticks"
plt.show()
```

```python
sns.kdeplot(titanic["Age"],shade=True) 
plt.xlabel("Age") 
sns.set_style("ticks") 
plt.show()
```

```python
sns.set_style("white")  #set style before you generate plot
sns.kdeplot(titanic["Age"],shade=True) 
sns.despine(bottom=True,left=True)  # sns.despine() removes spines.right and top removed default
plt.xlabel("Age") 
plt.show()
```

```python
g = sns.FacetGrid(titanic,col="Pclass",size=6) #seaborn.FacetGrid to choose column to display
g.map(sns.kdeplot,"Age",shade=True)  
sns.despine(left=True,bottom=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass")
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass",hue="Sex",size=3)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
g= sns.FacetGrid(titanic,col="Survived",row="Pclass",hue="Sex",size=3,palette="Set2")
g.map(sns.kdeplot,"Age",shade=True).add_legend()
sns.despine(left=True,bottom=True)
g.set(ylim=(0,0.04))
plt.savefig("titanic.png")
plt.show()
```
